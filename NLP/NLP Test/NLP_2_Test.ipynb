```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag, ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
import matplotlib.pyplot as plt
import spacy
```

```python
text = """
In the rugged terrain of the Western Ghats, Shivaji Maharaj, the founder of the Maratha Empire, was preparing for another day of strategic planning and military campaigns. Known for his innovative tactics and leadership, Shivaji Maharaj was determined to strengthen his kingdom and protect his people from external threats.

Shivaji Maharaj’s stronghold was the Raigad Fort, a formidable fortress perched high on a hilltop. The fort's strategic location allowed him to oversee the surrounding lands and defend against invaders. His trusted advisors and generals would gather in the fort's grand hall to discuss plans, share intelligence, and make critical decisions for the empire's expansion.

The evenings in Shivaji Maharaj's court were vibrant with activity. He would host feasts and gatherings to bolster the morale of his soldiers and celebrate victories. During these events, Shivaji Maharaj would address his people, sharing his vision for a prosperous and united Maratha Empire. His leadership inspired loyalty and courage among his subjects.

As night fell, Shivaji Maharaj would retire to his quarters, reflecting on the day's achievements and preparing for the challenges ahead. His dreams were filled with visions of a strong and resilient empire, where justice and valor would prevail.
"""
```

```python
# 1. Tokenization
sentences = sent_tokenize(text)
words = word_tokenize(text)
print("Tokenized Sentences:", sentences)
print("Tokenized Words:", words)
```

```python
# 2. Lowercasing
lowercased_text = text.lower()
print("Lowercased Text:\n", lowercased_text)
```

```python
# 3. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
removed_stopwords = [word for word in words if word not in filtered_words]
print("Filtered Words (Stopwords Removed):", filtered_words)
print("Removed Stopwords:", removed_stopwords)
```

```python
# 4. Punctuation Removal
cleaned_text = text.translate(str.maketrans('', '', string.punctuation))
print("Cleaned Text (Punctuation Removed):\n", cleaned_text)
```

```python
# 5. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("Stemmed Words:", stemmed_words[:10])
```

```python
# 6. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("Lemmatized Words:", lemmatized_words[:10])
```

```python
# 7. Named Entity Recognition (NER)
def ner(text):
  word_tokens = word_tokenize(text)
  word_pos = pos_tag(word_tokens)
  print(ne_chunk(word_pos))
ner(text)
```

```python
# 8. Part-of-Speech Tagging
pos_tagged = nltk.pos_tag(words)
print("Part-of-Speech Tagged Text (Sample):", pos_tagged[:10])
```

```python
# 9. Term Frequency (TF) of "Shivaji"
tf_shivaji = words.count("Shivaji")
print("Term Frequency of 'Shivaji':", tf_shivaji)
```

```python
# 10. Frequency Distribution
word_counts = Counter(filtered_words)
top_10_words = word_counts.most_common(10)
print("Top 10 Words:", top_10_words)
```

```python
words, counts = zip(*top_10_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='blue')
plt.title('Frequency Distribution of Top 10 Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 11. Token Count
token_count = len(word_counts)
print("Token Count:", token_count)
```

```python
# 12. Sentence Count
sentence_count = len(sentences)
print("Sentence Count:", sentence_count)
```

```python
# 13. Stopword Frequency
stopword_counts = Counter(removed_stopwords)
top_3_stopwords = stopword_counts.most_common(3)
print("Top 3 Stopwords:", top_3_stopwords)
```

```python
# 14. Longest Word
longest_word = max(words, key=len)
print("Longest Word:", longest_word, "Length:", len(longest_word))
```

```python
# 15. Word Cloud
wordcloud = WordCloud(width=800, height=400).generate(' '.join(filtered_words))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 16. Noun Identification
nouns = [word for word, pos in pos_tagged if pos.startswith('NN')]
print("Nouns:", nouns)
```

```python
# 17. Verb Identification
verbs = [word for word, pos in pos_tagged if pos.startswith('VB')]
print("Verbs:", verbs)
```

```python
# 18. Word Context for "empire"
empire_sentences = [sentence for sentence in sentences if "empire" in sentence]
print("Sentences with 'empire':", empire_sentences)
```

```python
# 19. Named Entity Categorization
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
entities = [(ent.text, ent.label_) for ent in doc.ents]
print(entities)
```

```python
# 20. Word Frequency Comparison
kingdom_count = filtered_words.count("kingdom")
feast_count = filtered_words.count("feast")
print("Kingdom Count:", kingdom_count, "Feast Count:", feast_count)
```
